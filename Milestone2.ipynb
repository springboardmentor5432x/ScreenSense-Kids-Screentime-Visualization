```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Enable inline plotting for Colab
%matplotlib inline

# Load Cleaned Dataset
df = pd.read_csv("/content/ScreenSense_updated.csv")

# Set style
sns.set(style="whitegrid")
```

# **Univariate Analysis**

```python
# Age Distribution
plt.figure(figsize=(6,4))
sns.countplot(
    x='Age_Group',
    hue='Age_Group',
    data=df,
    palette='muted',
    legend=False,
    order=['Kids', 'Pre-Teens', 'Teens']
)
plt.title("Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Count")

# Custom x-axis labels with age ranges
plt.xticks(
    ticks=[0, 1, 2],
    labels=['Kids (8–11)', 'Pre-Teens (12–14)', 'Teens (15–18)']
)

plt.show()
```

```python
# Gender Distribution
plt.figure(figsize=(5,4))
sns.countplot(x='Gender', data=df, hue='Gender', palette='Set2', legend=False)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Screen Time Category Count
plt.figure(figsize=(7,4))
sns.countplot(
    x='ScreenTime_Category',
    hue='ScreenTime_Category',
    data=df,
    palette="Set2",
    legend=False,
    order=['Low Usage', 'Moderate Usage', 'High Usage', 'Very High Usage']
)
plt.title("Screen Time Usage Categories")
plt.xlabel("Usage Category")
plt.ylabel("Count")

# Add screen time ranges with rotation and spacing
plt.xticks(
    ticks=[0, 1, 2, 3],
    labels=[
        'Low Usage\n(0.00–2.00 hrs)',
        'Moderate Usage\n(2.01–5.00 hrs)',
        'High Usage\n(5.01–8.00 hrs)',
        'Very High Usage\n(8.01–24.00 hrs)'
    ],
    rotation=0
)

plt.tight_layout()
plt.show()
```

```python
# Day Type Distribution
plt.figure(figsize=(6,4))
sns.countplot(
    x='Day_Type',
    data=df,
    hue='Day_Type',
    palette='Set2',
    legend=False
)
plt.title('Distribution of Day Type')
plt.xlabel('Day Type')
plt.ylabel('Count')
plt.show()
```

```python
# Device Type Usage
plt.figure(figsize=(6,4))
sns.countplot(x='Primary_Device', data=df, hue='Primary_Device', palette='Set2', legend=False)
plt.title("Primary Device Used by Kids")
plt.xlabel("Device Type")
plt.ylabel("Count")
plt.show()
```

```python
# Exceeded_Recommended_Limit
plt.figure(figsize=(5,4))
sns.countplot(x='Exceeded_Recommended_Limit', hue='Exceeded_Recommended_Limit', data=df, palette='coolwarm', legend=False)
plt.title("Exceeded Recommended Screen Time Limit")
plt.xlabel("Exceeded Limit (True/False)")
plt.ylabel("Count")
plt.show()
```

```python
# Educational vs Recreational Usage Categories
plt.figure(figsize=(7,4))
sns.countplot(
    x='EduRec_Category',
    hue='EduRec_Category',
    data=df,
    palette='Set2',
    legend=False,
    order=['Mostly Recreational', 'Balanced Usage', 'Mostly Educational']
)
plt.title("Educational vs Recreational Usage Categories")
plt.xlabel("Educational-to-Recreational Ratio Category")
plt.ylabel("Count")

# Add range labels for clarity
plt.xticks(
    ticks=[0, 1, 2],
    labels=[
        'Mostly Recreational\n(0.30–0.40)',
        'Balanced Usage\n(0.41–0.50)',
        'Mostly Educational\n(0.51–0.60)'
    ],
    rotation=0
)

plt.tight_layout()
plt.show()
```

```python
# Urban vs Rural Distribution
plt.figure(figsize=(5,4))
sns.countplot(x='Urban_or_Rural', hue='Urban_or_Rural', data=df, palette='pastel', legend=False)
plt.title("Urban vs Rural Distribution")
plt.xlabel("Location Type")
plt.ylabel("Count")
plt.show()
```

```python
# Health Impacts
health_cols = [col for col in df.columns if col.startswith("HealthImpact_")]

health_counts = df[health_cols].sum().sort_values(ascending=False)

plt.figure(figsize=(8,4))
sns.barplot(
    x=health_counts.index,
    y=health_counts.values,
    hue=health_counts.index,
    palette='Set2',
    legend=False
)
plt.title("Frequency of Health Impacts")

# Customize x-axis labels (make sure number of labels matches number of bars)
custom_labels = ['Poor Sleep', 'Eye Strain', 'Potential Risk', 'Anxiety', 'Healthy', 'Obesity Risk']
plt.xticks(ticks=range(len(custom_labels)), labels=custom_labels, rotation=45)

plt.xlabel("Health Impact")
plt.ylabel("Count")
plt.show()
```

**6. Health Impact Analysis**

```python
# Health Impact vs Age Group
import numpy as np
health_summary = df.groupby('Age_Group')[health_cols].mean()

plt.figure(figsize=(7,4))
sns.heatmap(health_summary, annot=True, cmap='coolwarm')
plt.title("Average Health Impact Scores by Age Group")

# Customize x-axis labels
custom_labels = ['Poor Sleep', 'Eye Strain', 'Potential Risk', 'Anxiety', 'Healthy', 'Obesity Risk']
plt.xticks(ticks=np.arange(len(custom_labels)) + 0.5, labels=custom_labels, rotation=90)

plt.xlabel("Health Impact Type")
plt.ylabel("Age Group")
plt.show()
```

```python
# Health Impact vs Gender
health_cols = [
    'HealthImpact_Anxiety', 'HealthImpact_Eye Strain', 'HealthImpact_Healthy',
    'HealthImpact_Obesity Risk', 'HealthImpact_Poor Sleep', 'HealthImpact_Potential Risk'
]

# Calculate average impact for each gender
health_gender = df.groupby('Gender')[health_cols].mean()

# Plot heatmap
plt.figure(figsize=(7,4))
sns.heatmap(health_gender, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Average Health Impact Scores by Gender")

# Customize x-axis labels
custom_labels = ['Poor Sleep', 'Eye Strain', 'Potential Risk', 'Anxiety', 'Healthy', 'Obesity Risk']
plt.xticks(ticks=np.arange(len(custom_labels)) + 0.5, labels=custom_labels, rotation=90)

plt.xlabel("Health Impact Type")
plt.ylabel("Gender")
plt.show()
```

```python
# Health Impact vs Exceeded Recommended Limit
plt.figure(figsize=(7,4))
health_cols = [
    'HealthImpact_Anxiety', 'HealthImpact_Eye Strain', 'HealthImpact_Healthy',
    'HealthImpact_Obesity Risk', 'HealthImpact_Poor Sleep', 'HealthImpact_Potential Risk'
]

# Compute average values grouped by Exceeded_Recommended_Limit
health_exceeded = df.groupby('Exceeded_Recommended_Limit')[health_cols].mean()

# Plot heatmap
sns.heatmap(
    health_exceeded,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Average Health Impact Scores by Exceeded Recommended Limit')

# Customize x-axis labels
custom_labels = ['Poor Sleep', 'Eye Strain', 'Potential Risk', 'Anxiety', 'Healthy', 'Obesity Risk']
plt.xticks(ticks=np.arange(len(custom_labels)) + 0.5, labels=custom_labels, rotation=90)

plt.xlabel('Health Impact Type')
plt.ylabel('Exceeded Recommended Limit')
plt.show()
```

```python
# Health Impact vs Location
plt.figure(figsize=(7,4))

health_cols = [
    'HealthImpact_Anxiety', 'HealthImpact_Eye Strain', 'HealthImpact_Healthy',
    'HealthImpact_Obesity Risk', 'HealthImpact_Poor Sleep', 'HealthImpact_Potential Risk'
]

# Calculate average values grouped by Urban_or_Rural
health_location = df.groupby('Urban_or_Rural')[health_cols].mean()

# Plot heatmap
sns.heatmap(
    health_location,
    annot=True,
    cmap='YlOrRd',
    fmt='.2f'
)

plt.title('Average Health Impact Scores by Location Type')

# Customize x-axis labels
custom_labels = ['Poor Sleep', 'Eye Strain', 'Potential Risk', 'Anxiety', 'Healthy', 'Obesity Risk']
plt.xticks(ticks=np.arange(len(custom_labels)) + 0.5, labels=custom_labels, rotation=90)

plt.xlabel('Health Impact Type')
plt.ylabel('Location (Urban / Rural)')
plt.show()
```
