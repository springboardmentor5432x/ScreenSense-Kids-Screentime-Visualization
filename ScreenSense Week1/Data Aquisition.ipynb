```python
#Unique Value per categorical(where datatype is object) column
for col in df.select_dtypes(include='object').columns:
    print(f"{col}:{df[col].nunique()} Unique Values")
```

```python
#Visualisations
import matplotlib.pyplot as plt
```

```python
#histogram
plt.hist(df['Avg_Daily_Screen_Time_hr'],bins=10,color='orange',edgecolor='black')
plt.title('Distribution of Kids Daily Screen time hours')
plt.xlabel('Screen Time (Hours)')
plt.ylabel('No. of kids')
plt.show()
```

```python
#pie chart
screen_time_by_location = df.groupby('Urban_or_Rural')['Avg_Daily_Screen_Time_hr'].sum()
plt.pie(screen_time_by_location,
        labels=screen_time_by_location.index,
        colors=['green','orange'])
plt.title('Urban vs Rural by screentime')
plt.show()
```

```python
device_counts = df['Primary_Device'].value_counts() #no. of children using each device type
plt.pie(device_counts,
        labels=device_counts.index,
        colors=['skyblue','yellow','blue','brown'])
plt.title('Most commonly used device among children')
plt.show()
```
