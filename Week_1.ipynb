```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset

file_path = "D:\Indian_Kids_Screen_Time.csv"
df = pd.read_csv(file_path)
```

```python
# Basic overview

print(" Dataset Loaded Successfully!\n")
print("Total Rows:", df.shape[0])
print("Total Columns:", df.shape[1])
```

```python
# Explore Schema and Data Types

print("\n Dataset Information:\n")
df.info()
```

```python
print("\n First 5 Rows:\n")
display(df.head())
```

```python
print("\n Statistical Summary (Numeric Columns):\n")
display(df.describe())
```

```python
print("\n Summary (All Columns):\n")
display(df.describe(include='all'))
```

```python
# Check Missing Values and Duplicates

print("\n Missing Values per Column:\n")
print(df.isnull().sum())
```

```python
print("\n Duplicate Rows:", df.duplicated().sum())
```

```python
# 5Ô∏è Unique Values per Column

print("\n Unique Values per Column:\n")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
# Data Type Overview

print("\n Data Types Summary:")
print(df.dtypes)
```

```python
# Basic Visual Checks

# Distribution of Average Daily Screen Time
plt.figure(figsize=(8,5))
sns.histplot(df['Avg_Daily_Screen_Time_hr'], kde=True, bins=30, color='skyblue')
plt.title("Distribution of Average Daily Screen Time (hours)")
plt.xlabel("Avg Daily Screen Time (hours)")
plt.ylabel("Count")
plt.show()
```

```python
# Count by Gender
plt.figure(figsize=(5,4))
sns.countplot(data=df, x='Gender', palette='pastel')
plt.title("Gender Distribution")
plt.show()
```

```python
# Urban vs Rural
plt.figure(figsize=(5,4))
sns.countplot(data=df, x='Urban_or_Rural', palette='Set2')
plt.title("Urban vs Rural Distribution")
plt.show()
```
