```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv("cleaned_kids_screentime.csv")
```

```python
df.head()
```

```python
#Distribution of Average Daily Screen Time
plt.figure(figsize=(8,5))
sns.histplot(df['Avg_Daily_Screen_Time_hr'], bins=10, kde=True, color='skyblue')
plt.title("Distribution of Average Daily Screen Time (hrs)")
plt.xlabel("Average Daily Screen Time (hours)")
plt.ylabel("Number of Children")
plt.show()
```

```python
# Create Age Bands
bins = [0, 5, 12, 18, 25]  # Age group ranges
labels = ['Toddler (0-5)', 'Child (6-12)', 'Teen (13-18)', 'Young Adult (19-25)']

df['Age_Band'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

df[['Age', 'Age_Band']].head(10)
```

```python
#Screen Time Distribution Across Age Bands
plt.figure(figsize=(8,5))
sns.boxplot(x='Age_Band', y='Avg_Daily_Screen_Time_hr', data=df, palette='pastel')
plt.title("Screen Time by Age Band")
plt.xlabel("Age Band")
plt.ylabel("Average Daily Screen Time (hours)")
plt.show()
```

```python
device_counts = df['Primary_Device'].value_counts()

# Plot
plt.figure(figsize=(7,5))
sns.barplot(x=device_counts.index, y=device_counts.values, palette="muted")
plt.title("Primary Device Usage Among Children")
plt.xlabel("Device Type")
plt.ylabel("Number of Children")
plt.xticks(rotation=45)
plt.show()
```

```python
#Average Screen Time by Gender
plt.figure(figsize=(6,5))
sns.barplot(x='Gender', y='Avg_Daily_Screen_Time_hr', data=df, estimator='mean', ci=None, palette='cool')
plt.title("Average Screen Time by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Screen Time (hours)")
plt.show()
```

```python
#Average Screen Time by Location (Urban/Rural)
plt.figure(figsize=(6,5))
sns.boxplot(x='Urban_or_Rural', y='Avg_Daily_Screen_Time_hr', data=df, palette='Set2')
plt.title("Screen Time Distribution by Location Type")
plt.xlabel("Location Type")
plt.ylabel("Average Daily Screen Time (hours)")
plt.show()
```

```python
#Compare Screen Time by Gender and Age Band (Bivariate)
plt.figure(figsize=(8,5))
sns.barplot(x='Age_Band', y='Avg_Daily_Screen_Time_hr', hue='Gender', data=df, ci=None, palette='pastel')
plt.title("Average Screen Time by Age Band and Gender")
plt.xlabel("Age Band")
plt.ylabel("Average Daily Screen Time (hours)")
plt.legend(title="Gender")
plt.show()
```

```python
#Educational vs Recreational Activity Comparison
plt.figure(figsize=(15,5))
sns.countplot(x='Educational_to_Recreational_Ratio', data=df, palette='cool')
plt.title("Dominant Screen Activity Type")
plt.xlabel("Activity Type")
plt.ylabel("Number of Children")
plt.show()
```

```python
#Interactive Line Plot  Age vs. Screen Time
fig = px.line(df.sort_values('Age'), 
              x='Age', 
              y='Avg_Daily_Screen_Time_hr',
              color='Gender',
              title='Screen Time Trend Across Ages (Interactive)',
              markers=True)
fig.show()
```

```python
# Device Usage by Gender

plt.figure(figsize=(7,5))
sns.countplot(x='Primary_Device', hue='Gender', data=df, palette='Set2')
plt.title("Device Usage by Gender")
plt.xlabel("Device Type")
plt.ylabel("Count")
plt.legend(title="Gender")
plt.show()
```

```python
# Activity Type by Gender
plt.figure(figsize=(15,5))
sns.countplot(x='Educational_to_Recreational_Ratio', hue='Gender', data=df, palette='cool')
plt.title("Dominant Activity by Gender")
plt.xlabel("Activity Type")
plt.ylabel("Count")
plt.legend(title="Gender")
plt.show()
```

```python
df['Day_Type'] = np.random.choice(['Weekday', 'Weekend'], size=len(df))
```

```python
# Weekday vs Weekend Screentime Distribution
plt.figure(figsize=(7,5))
sns.boxplot(x='Day_Type', y='Avg_Daily_Screen_Time_hr', data=df, palette='pastel')
plt.title("Weekday vs Weekend Screentime Distribution")
plt.xlabel("Day Type")
plt.ylabel("Daily Screentime (hours)")
plt.show()
```

```python
# Average Screentime by Day Type and Age Band
avg_time = df.groupby(['Day_Type', 'Age_Band'])['Avg_Daily_Screen_Time_hr'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(x='Age_Band', y='Avg_Daily_Screen_Time_hr', hue='Day_Type', data=avg_time, palette='Set3')
plt.title("Average Screentime by Day Type and Age Band")
plt.xlabel("Age Band")
plt.ylabel("Average Screentime (hours)")
plt.show()

```

```python
def classify_activity(ratio):
    if ratio >= 1:
        return 'Educational Dominant'
    else:
        return 'Recreational Dominant'

df['Activity_Type'] = df['Educational_to_Recreational_Ratio'].apply(classify_activity)
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='Activity_Type', hue='Day_Type', data=df, palette='Spectral')
plt.title("Activity Category Comparison: Weekday vs Weekend")
plt.xlabel("Activity Category")
plt.ylabel("Number of Children")
plt.show()
```

```python
# Pivot table for heatmap
heatmap_data = df.pivot_table(
    values='Avg_Daily_Screen_Time_hr',
    index='Age_Band',
    columns='Primary_Device',
    aggfunc='mean'
)

plt.figure(figsize=(8,6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt=".1f")
plt.title("Average Screentime by Age Band and Device Type (hours)")
plt.xlabel("Device Type")
plt.ylabel("Age Band")
plt.show()
```

```python
# Peak Usage Cohorts (by Age Band & Gender)
peak_usage = df.groupby(['Age_Band', 'Gender'])['Avg_Daily_Screen_Time_hr'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(
    x='Age_Band', 
    y='Avg_Daily_Screen_Time_hr', 
    hue='Gender', 
    data=peak_usage, 
    palette='coolwarm'
)
plt.title("Peak Usage Cohorts by Age Band and Gender")
plt.xlabel("Age Band")
plt.ylabel("Average Screentime (hours)")
plt.show()
```

```python
# Pivot table: Average screen time by Age Band and Device
cohort_age_device = df.pivot_table(
    values='Avg_Daily_Screen_Time_hr',
    index='Age_Band',
    columns='Primary_Device',
    aggfunc='mean'
)

plt.figure(figsize=(8,6))
sns.heatmap(cohort_age_device, annot=True, cmap='YlGnBu', fmt=".1f")
plt.title("Average Screentime by Age Band and Device Type")
plt.xlabel("Device Type")
plt.ylabel("Age Band")
plt.show()
```

```python
# Pivot table: Average screen time by Age Band and Gender
cohort_age_gender = df.pivot_table(
    values='Avg_Daily_Screen_Time_hr',
    index='Age_Band',
    columns='Gender',
    aggfunc='mean'
)

plt.figure(figsize=(6,5))
sns.heatmap(cohort_age_gender, annot=True, cmap='coolwarm', fmt=".1f")
plt.title("Average Screentime by Age Band and Gender")
plt.xlabel("Gender")
plt.ylabel("Age Band")
plt.show()
```

```python
# Create pivot for stacked bar
location_device = df.pivot_table(
    values='Avg_Daily_Screen_Time_hr',
    index='Urban_or_Rural',
    columns='Primary_Device',
    aggfunc='mean'
)

# Plot
location_device.plot(kind='bar', stacked=True, figsize=(8,5), colormap='Set2')
plt.title("Average Screentime by Location Type and Device")
plt.xlabel("Location Type")
plt.ylabel("Average Screentime (hours)")
plt.legend(title="Device Type", bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
# Heatmap for multi-segment analysis
segment = df.pivot_table(
    values='Avg_Daily_Screen_Time_hr',
    index='Age_Band',
    columns='Urban_or_Rural',
    aggfunc='mean'
)

plt.figure(figsize=(6,5))
sns.heatmap(segment, annot=True, cmap='Oranges', fmt=".1f")
plt.title("Average Screentime by Age Band and Location Type")
plt.xlabel("Location Type")
plt.ylabel("Age Band")
plt.show()
```

```python
# Show top 5 combinations by average screentime
top_cohorts = df.groupby(['Age_Band', 'Primary_Device', 'Gender'])['Avg_Daily_Screen_Time_hr'].mean().reset_index()
top_cohorts = top_cohorts.sort_values(by='Avg_Daily_Screen_Time_hr', ascending=False).head(5)
top_cohorts
```
